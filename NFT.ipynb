```python
from selenium import webdriver
import pandas as pd
import re
```

```python
crypto_names = []
crypto_ranks = []

driver = webdriver.Chrome('/usr/bin/chromedriver')
for i in range(1,41):
    driver.get('https://www.stateofthedapps.com/rankings/platform/ethereum?page=' + str(i) )
    names = driver.find_elements_by_class_name('name')
    ranks = driver.find_elements_by_class_name('rank')
    
    for name in names:    
        crypto_names.append(name.text)
        
    for rank in ranks:
        crypto_ranks.append(rank.text)
        
    
    
    
```

```python
new_crypto_names = []

for name in crypto_names:
    name = str.lower(name)
    if '-' in name:
        name = name.replace('-',' ')  
        name = ' '.join(name.split())
    name = name.replace(' ','-')
    new_crypto_names.append(name)
        

        
        
    
```

```python
crypto_nft = []
crypto_description = []



driver = webdriver.Chrome('/usr/bin/chromedriver')
for i in new_crypto_names[:100]:
    driver.get('https://www.stateofthedapps.com/dapps/' + str.lower(i))
    driver.implicitly_wait(10)    
    crypto_nft.append(driver.find_element_by_class_name('category-link').text)
    crypto_description.append(driver.find_element_by_class_name('description').text)

        
    
    
    
    
    
    
```

```python
for i in range(len(crypto_description)):
    crypto_description[i] = crypto_description[i].replace('\n',' ')

for j in range(len(crypto_ranks[:100])):
    if '\n' in crypto_ranks[j]:
        crypto_ranks[j] = crypto_ranks[j][:crypto_ranks[j].index('\n')]
```

```python
data = pd.DataFrame({'Name' : new_crypto_names[:100] , 'Description' : crypto_description , 'NFT' : crypto_nft , 'Rank' : crypto_ranks[:100]})
```
